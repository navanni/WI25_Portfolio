Recent research highlights the growing impact of environmental, social, and governance (ESG) factors on firm performance. Their study provides evidence that ESG-conscious companies may have a competitive advantage, but it does not specifically address performance in high-volatility market environments. Similarly, Albuquerque, Koskinen, and Zhang (2020) argue that firms with strong ESG commitments experience lower systematic risk and tend to perform better during downturns. Aydoğmuş et al (2022) concludes that ESG performance (especially the social and governance aspects) has a significant, positive relationship with profitability and firm value. 

While it did not find a significant relationship with the environmental aspect, they suggested that environmental actions may take longer to affect profitability because of slower progress and higher costs. These findings suggest that companies with strong environmental risk management might be more resilient during volatile market conditions, aligning with our research focus.

**SP 500 ESG Risk Ratings** is a dataset with each of the S&P 500 companies and their corresponding Environmental, Social, and Governance (ESG) scores. ESG Risk Ratings are a numerical rating from 0-100 that is evaluate based on environmental risks: carbon footprint and greenhouse gas emissions, water usage, waste management, climate change risk, etc, social risks: labor practices, diversity and inclusion, human rights, community engagement, etc, and governance risks: executive compensation, business ethics and transparency, regulatory compliance, etc. Scores in each category are added together as weighted averages to form the Total ESG Risk Score. A higher rating indicates a better management of theses risks and therefore more sustainable and ethical practices, while a lower rating suggests potential shortcomings or poor handling of ESG-related issues. The dataset includes a lot of other information that we do not need, so there are only a few columns that we care about for this project, ‘Name’ (of the company) and ‘Total ESG Risk Score', and the disaggregated values 'Environment Risk Score', 'Governance Risk Score', and 'Social Risk Score'. 


**The CBOE VIX dataset** is using the Chicago Board Option Exchange for the Volatility Index, which is a measure of implied volatility and reflects how much investors anticipate the market to fluctuate in the near future (next ~30 days). Generally, a VIX value of over 30 is linked to a larger volatility and therefore increased uncertainty, investors’ fear, and risk while values below 20 correspond to more stable periods in the market. VIX is impacted by a variety of factors, including geopolitical events, market dynamics, and economic indicators (like GDP, inflation, interest rates, consumer spending, etc) as well as investor sentiment (eg positive market news can reduce fear, lowering the VIX), trading volumes, and option prices. The dataset shows what the VIX was at the beginning of each month from May 2020 to May 2023 (we will disregard the other columns besides ‘Date’ and ‘Open’). 

**The Sharpe Ratio dataset** shows the measure of the risk-adjusted return of an investment and helps to understand how well the return of an asset compensates for the risk taken. It is calculated with a formula that takes into account the average return of the investment, the risk-free rate (based on 3-Month Treasury Bill), and the standard deviation of the investment’s return. A high Sharpe Ratio (>1) would mean that the investment is providing a better return for the level of risk taken, which is generally preferred. A low Sharpe Ratio  (<1, or negative) means that the returns do not justify the level of risk and that the investment may not be a good choice. This is a useful way to compare different investments with varying levels of risk and return and allow investors to have more information when making decisions.

## Section 1: Testing for Correlations between Different Variables of the ESG and the AVG_Sharpe Ratio 

The below preliminary correlation check revealed that Environment, Social, and Governance scores are moderately correlated with one another, suggesting they might collectively influence risk‐adjusted returns more effectively than focusing on a single dimension. Based on these findings, we decided to **shift our question** from focusing solely on Environmental factors to using a combined Total ESG measure; the rest of our EDA will therefore focus on how average ESG risk relates to VIX and Sharpe ratios, rather than isolating the environmental component.








```python

# 1. READ THE S&P 500 ESG DATA (KEEPING ALL ROWS)

esg_file = "SP 500 ESG Risk Ratings.csv"
risk_ratings = pd.read_csv(esg_file)

# 2. READ THE SHARPE RATIO DATA (KEEPING ALL ROWS)
sharpe_file = "Sharpe Ratios for S&P 500 Co.csv"
df_sharpe = pd.read_csv(sharpe_file)

# 3. MERGE THE TWO DATASETS
merged_df = pd.merge(
    risk_ratings, 
    df_sharpe, 
    on="Symbol",
    how="inner"    
)

print("Merged dataset preview:")
display(merged_df.head())

# 4. ENCODE SECTOR TO INCLUDE IT IN THE CORRELATION MATRIX

if "Sector" in merged_df.columns:
    merged_df["Sector"] = merged_df["Sector"].astype(str)
    sectors = merged_df["Sector"].unique()
    sec_dict = {sec: i for i, sec in enumerate(sectors)}
    reverse_sec_dict = {i: sec for sec, i in sec_dict.items()}
    merged_df["Sector_Encoded"] = merged_df["Sector"].map(sec_dict)
else:
    print("No 'Sector' column found to encode.")

# 5. CREATE AN AVERAGE SHARPE RATIO COLUMN

sharpe_cols = ["Sept 22", "April 22", "Feb 22", "Jan 21", "Oct 20", "June 20"]
existing_sharpe_cols = [c for c in sharpe_cols if c in merged_df.columns]

if existing_sharpe_cols:
    merged_df["Avg_Sharpe"] = merged_df[existing_sharpe_cols].mean(axis=1, skipna=True)
else:
    print("No matching Sharpe columns found in the merged dataset.")

# 6. BUILD A CORRELATION MATRIX

num_cols = []
for col in [
    "Environment Risk Score",
    "Social Risk Score",
    "Governance Risk Score",
    "Total ESG Risk Score",
    "Avg_Sharpe",
    "Sector_Encoded"
]:
    if col in merged_df.columns:
        num_cols.append(col)

# Drop rows with missing data in these columns
df_for_corr = merged_df[num_cols].dropna()
corr_matrix = df_for_corr.corr()

print("\nCorrelation Matrix:")
display(corr_matrix)

# 7. VISUALIZE THE CORRELATION MATRIX WITH ANNOTATIONS INSIDE THE CELLS
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix for ESG Scores, Sector, and Sharpe Ratios")
plt.xticks(rotation=45, ha="right")
plt.show()

```

```python
sec = 'Sector_Encoded'
cats = df_for_corr[sec].unique()
corr_matrices = {}

for cat in cats:
    cat_df = df_for_corr[df_for_corr[sec] == cat]
    cat_df = cat_df.drop(columns = [sec])
    corr_matrices[cat] = cat_df.corr()

for cat, mat in corr_matrices.items():
    plt.figure(figsize=(8, 6))
    sns.heatmap(mat, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)

    plt.title(f"Correlation Matrix for ESG Scores, Sector {reverse_sec_dict[cat]}, and Sharpe Ratios")
    plt.xticks(rotation=45, ha="right")
    plt.show()
```

```python
df = final_timeseries.copy()

# 1. Create bins for VIX. Split into four quartiles (Low, Med-Low, Med-High, High)
df['VIX_bin'] = pd.qcut(df['Close'], q=4, labels=['Low Vol','Med-Low Vol','Med-High Vol','High Vol'])

# 2. Group by Sector and the new VIX_bin
grouped = df.groupby(['Sector', 'VIX_bin'])

for (sector, vol_bin), group_data in grouped:
    # Compute correlation between Sharpe Ratio and ESG in this subset
    corr_value = group_data[['Sharpe Ratio', 'Total ESG Risk score']].corr().iloc[0,1]
    
    print(f"Sector: {sector} | Volatility Bin: {vol_bin} | Correlation (Sharpe vs ESG): {corr_value:.2f}")
```

```python
# Create a grid of values for VIX (Close) and pick a few representative ESG scores
vix_range = np.linspace(df['Close'].min(), df['Close'].max(), 50)

# Use three ESG levels: 25th, 50th, and 75th percentiles
esg_levels = [
    df['Total ESG Risk score'].quantile(0.25),
    df['Total ESG Risk score'].quantile(0.50),
    df['Total ESG Risk score'].quantile(0.75)
]

plot_data = []
for esg in esg_levels:
    for vix in vix_range:
        plot_data.append({
            'Close': vix,
            'Total ESG Risk score': esg
        })

plot_df = pd.DataFrame(plot_data)

# Get the model's predicted Sharpe Ratios for each (VIX, ESG) pair
plot_df['predicted_sharpe'] = model.predict(plot_df)

# Plot lines for each ESG level
plt.figure(figsize=(8, 6))
for esg in esg_levels:
    subset = plot_df[plot_df['Total ESG Risk score'] == esg]
    plt.plot(subset['Close'], subset['predicted_sharpe'], label=f'ESG={esg:.2f}')

plt.xlabel('VIX (Close)')
plt.ylabel('Predicted Sharpe Ratio')
plt.title('Predicted Sharpe Ratio by ESG Score and VIX')
plt.legend(title='ESG Level')
plt.show()
```
