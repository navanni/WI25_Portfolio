# ESG Factors and Financial Performance

Recent research highlights the growing impact of environmental, social, and governance (ESG) factors on firm performance. Khan, Serafeim, and Yoon (2016) found that firms with high sustainability scores tend to exhibit superior risk-adjusted returns, particularly when ESG factors are material to the industry. Their study provides evidence that ESG-conscious companies may have a competitive advantage, but it does not specifically address performance in high-volatility market environments. Similarly, Albuquerque, Koskinen, and Zhang (2020) argue that firms with strong ESG commitments experience lower systematic risk and tend to perform better during downturns. Aydoğmuş et al (2022) concludes that ESG performance (especially the social and governance aspects) has a significant, positive relationship with profitability and firm value. 

While it did not find a significant relationship with the environmental aspect, they suggested that environmental actions may take longer to affect profitability because of slower progress and higher costs. These findings suggest that companies with strong environmental risk management might be more resilient during volatile market conditions, aligning with our research focus.

# ESG Performance and Market Volatility (VIX > 30)

Other studies explore the relationship between ESG factors and market turbulence. Engle, Giglio, Kelly, Lee, and Stroebel (2020) demonstrated that climate-conscious firms are less exposed to downside risk, suggesting that ESG factors could serve as a hedge against market volatility. Meanwhile, Brière, Peillex, and Ureche-Rangau (2019) analyzed whether socially responsible investing (SRI) funds outperform during turbulent market conditions. Their results indicate that high-ESG funds tend to be more resilient but may underperform in certain volatility regimes, raising the question of whether environmental risk scores specifically influence risk-adjusted returns during extreme volatility periods. 

While these studies provide insights into ESG and volatility, they do not directly measure the interaction between environmental risk scores and stock performance under extreme volatility (VIX > 30), which is the central focus of our research.

**SP 500 ESG Risk Ratings** is a dataset with each of the S&P 500 companies and their corresponding Environment Risk Score. Environment Risk Score is a numerical rating from 0-100 that indicates how much a company or its activities harm the environment—a higher score corresponds to a greater risk of negative impact while a lower score indicates better environmental practices, taking factors like water usage and disposal, greenhouse gas emissions, biodiversity impact, and compliance with environmental regulations into account. The dataset includes a lot of other information that we do not need, so there are only two columns that we care about for this project, ‘Name’ (of the company) and ‘Environment Risk Score. 

**The CBOE VIX dataset** is using the Chicago Board Option Exchange for the Volatility Index, which is a measure of implied volatility and reflects how much investors anticipate the market to fluctuate in the near future (next ~30 days). Generally, a VIX value of over 30 is linked to a larger volatility and therefore increased uncertainty, investors’ fear, and risk while values below 20 correspond to more stable periods in the market. VIX is impacted by a variety of factors, including geopolitical events, market dynamics, and economic indicators (like GDP, inflation, interest rates, consumer spending, etc) as well as investor sentiment (eg positive market news can reduce fear, lowering the VIX), trading volumes, and option prices. The dataset shows what the VIX was at the close of each month from May 2020 to May 2023 (we will disregard the other columns besides ‘Date’ and ‘Close’). 

**The Sharpe Ratio dataset** shows the measure of the risk-adjusted return of an investment and helps to understand how well the return of an asset compensates for the risk taken. It is calculated with a formula that takes into account the average return of the investment, the risk-free rate (based on 3-Month Treasury Bill), and the standard deviation of the investment’s return. A high Sharpe Ratio (>1) would mean that the investment is providing a better return for the level of risk taken, which is generally preferred. A low Sharpe Ratio  (<1, or negative) means that the returns do not justify the level of risk and that the investment may not be a good choice. This is a useful way to compare different investments with varying levels of risk and return and allow investors to have more information when making decisions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from IPython.display import display
import seaborn as sns

# 1. READ THE S&P 500 ESG DATA (KEEPING ALL ROWS)

esg_file = "SP 500 ESG Risk Ratings.csv"
risk_ratings = pd.read_csv(esg_file)

# 2. READ THE SHARPE RATIO DATA (KEEPING ALL ROWS)
sharpe_file = "Sharpe Ratios for S&P 500 Co.csv"
df_sharpe = pd.read_csv(sharpe_file)

# 3. MERGE THE TWO DATASETS
merged_df = pd.merge(
    risk_ratings, 
    df_sharpe, 
    on="Symbol",   # Change to 'Name' if that's the common field in your case
    how="inner"    # Use a different merge type if necessary
)

print("Merged dataset preview:")
display(merged_df.head())

# 4. ENCODE SECTOR TO INCLUDE IT IN THE CORRELATION MATRIX

if "Sector" in merged_df.columns:
    le = LabelEncoder()
    merged_df["Sector"] = merged_df["Sector"].astype(str)  
    merged_df["Sector_Encoded"] = le.fit_transform(merged_df["Sector"])
else:
    print("No 'Sector' column found to encode.")

# 5. CREATE AN AVERAGE SHARPE RATIO COLUMN

sharpe_cols = ["Sept 22", "April 22", "Feb 22", "Jan 21", "Oct 20", "June 20"]
existing_sharpe_cols = [c for c in sharpe_cols if c in merged_df.columns]

if existing_sharpe_cols:
    merged_df["Avg_Sharpe"] = merged_df[existing_sharpe_cols].mean(axis=1, skipna=True)
else:
    print("No matching Sharpe columns found in the merged dataset.")

# 6. BUILD A CORRELATION MATRIX

num_cols = []
for col in [
    "Environment Risk Score",
    "Social Risk Score",
    "Governance Risk Score",
    "Total ESG Risk Score",
    "Avg_Sharpe",
    "Sector_Encoded"
]:
    if col in merged_df.columns:
        num_cols.append(col)

# Drop rows with missing data in these columns
df_for_corr = merged_df[num_cols].dropna()
corr_matrix = df_for_corr.corr()

print("\nCorrelation Matrix:")
display(corr_matrix)

# 7. VISUALIZE THE CORRELATION MATRIX WITH ANNOTATIONS INSIDE THE CELLS
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix for ESG Scores, Sector, and Sharpe Ratios")
plt.xticks(rotation=45, ha="right")
plt.show()

```
